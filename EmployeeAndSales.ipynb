```python
import pandas as pd
df = pd.read_csv('/kaggle/input/employee-performance-and-salary-dataset/Employe_Performance_dataset.csv')
print("# Task 1: Display the first 10 rows of the dataset")
print(df.head(10))
```

```python
print("# Task 2: Column names and their respective data types")
print(df.dtypes)
```

```python
print("# Task 3: Number of rows and columns")
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
```

```python
print("# Task 4: Select Name, Age, and Salary columns")
print(df[['Name', 'Age', 'Salary']])
```

```python
print("# Task 5: Select the first 5 rows using iloc")
print(df.iloc[:5])
```

```python
print("# Task 6: Select all rows where the Department is 'Sales'")
print(df.loc[df['Department'] == 'Sales'])
```

```python
print("# Task 7: Select rows where Salary is greater than 8000 and Status is 'Active'")
print(df.loc[(df['Salary'] > 8000) & (df['Status'] == 'Active')])
```

```python
print("# Task 8: Extract rows where Experience > 5 years and Performance Score >= 4")
print(df.loc[(df['Experience'] > 5) & (df['Performance Score'] >= 4)])
```

```python
print("# Task 9: Select every alternate row from the DataFrame")
print(df.iloc[::2])
```

```python
salary_threshold = df['Salary'].quantile(0.9)
print("# Task 10: Retrieve rows where employee belongs to 'Sales' and has Salary in the top 10%")
print(df.loc[(df['Department'] == 'Sales') & (df['Salary'] >= salary_threshold)])
```

```python
print("# Task 11: Replace all Inactive values in the Status column with Retired")
df['Status'] = df['Status'].replace("Inactive", "Retired")
print(df['Status'].unique())
```

```python
print("# Task 12: Add a new column Bonus where values are 10% of the Salary column.")
df['Bonus'] = df['Salary'] * 0.1
print(df.head())
```

```python
print("# Task 14: Update the Performance Score of employees in the IT department to 5 if it is NaN")
df.loc[(df['Department'] == "IT") & (df['Performance Score'].isna()), 'Performance Score'] = 5
print(df.head())
```

```python
print("# Task 14: Modify the Experience of employees with less than 2 years to 2.")
df.loc[df['Experience'] < 2, 'Experience'] = 2
print(df['Experience'].unique())
```

```python
print("# Task 15: Normalize the Salary column so that all values are between 0 and 1")
df['Salary_Normalized'] = (df['Salary'] - df['Salary'].min()) / (df['Salary'].max() - df['Salary'].min())
print(df[['Salary', 'Salary_Normalized']].head())
```

```python
print("# Task 16: Replace all Location values with Remote for employees with a Performance Score below 3.")
df.loc[df['Performance Score'] < 3, 'Location'] = "Remote"
print(df['Location'].unique())
```

```python
print("# Task 17: Create a new column Experience Level with values Junior, Mid, or Senior based on the Experience.")
def experience_level(exp):
    if exp < 3:
        return "Junior"
    elif exp < 7:
        return "Mid"
    else:
        return "Senior"

df['Experience Level'] = df['Experience'].apply(experience_level)
print(df[['Experience', 'Experience Level']].head())
```

```python
print("# Task 18: Add a new row with specific values using the append() method")
import pandas as pd

new_row = {
    'Name': 'John Doe', 
    'Age': 30, 
    'Salary': 7000, 
    'Department': 'HR', 
    'Experience': 5, 
    'Performance Score': 4, 
    'Status': 'Active',
    'Location': 'New York'
}
new_row_df = pd.DataFrame([new_row])  
df = pd.concat([df, new_row_df], ignore_index=True)
print(df.tail())
```

```python
print("# Task 19: Delete the column Session from the DataFrame.")
if 'Session' in df.columns:
    df = df.drop(columns=['Session'])
print(df.columns)
```

```python
print("# Task 20: Delete rows where the Age is greater than 60")
df = df[df['Age'] <= 60]
print(df['Age'].max())
```

```python
print("# Task 21: Add a new column Age Group with values Young, Middle-aged, or Senior based on Age")
def age_group(age):
    if age < 30:
        return "Young"
    elif age <= 50:
        return "Middle-aged"
    else:
        return "Senior"

df['Age Group'] = df['Age'].apply(age_group)
print(df[['Age', 'Age Group']].head())
```

```python
print("# Task 22: Identify all columns with missing values using the isna() method")
missing_values = df.isna().sum()
print("Missing Values per Column:")
print(missing_values)
```

```python
print("# Task 23:  Fill the missing values in Performance Score with the average score of that column")
df['Performance Score'] = df['Performance Score'].fillna(df['Performance Score'].mean())
print(df['Performance Score'].isna().sum())
```

```python
print("# Task 24: Drop all rows with missing values using the dropna() method")
df_cleaned = df.dropna()
print(df_cleaned.isna().sum())
```

```python
print("# Task 25: Replace missing values in Performance Score with the median of the scores, grouped by Department")
df['Performance Score'] = df.groupby('Department')['Performance Score'].transform(
    lambda x: x.fillna(x.median())
)
print(df['Performance Score'].isna().sum())
```

```python
print("# Task 26: Identify rows where the employee is in the IT department and their Age is above the average Age of IT employees")
it_avg_age = df.loc[df['Department'] == "IT", 'Age'].mean()
it_above_avg_age = df.loc[(df['Department'] == "IT") & (df['Age'] > it_avg_age)]
print(it_above_avg_age)
```

```python
print("# Task 27: Identify rows with duplicate Name and Joining Date combinations, and remove duplicates")
if 'Joining Date' in df.columns:
    df = df.drop_duplicates(subset=['Name', 'Joining Date'], keep='first')
print(df.duplicated(subset=['Name', 'Joining Date']).sum())
```

```python
print("# Task 28: Remove rows where the Experience value is inconsistent with the Age (e.g., Experience > Age - 18)")
df = df[df['Experience'] <= (df['Age'] - 18)]
print(df)
```

```python
print("# Task 29: Sort the DataFrame by Salary in descending order")
df_sorted_salary = df.sort_values(by='Salary', ascending=False)
print(df_sorted_salary.head())
```

```python
print("# Task 30: Sort the DataFrame by Department and then by Joining Date in ascending order")
if 'Joining Date' in df.columns:
    df_sorted_dept_join = df.sort_values(by=['Department', 'Joining Date'], ascending=True)
    print(df_sorted_dept_join.head())
```

```python
print("# Task 31: Rank all employees by Salary within their Department and add the rank as a new column")
df['Salary Rank'] = df.groupby('Department')['Salary'].rank(ascending=False)
print(df[['Department', 'Salary', 'Salary Rank']].head())
```

```python
print("# Task 32:  Sort the dataset by Experience in descending order, but break ties using Performance Score in ascending order")
df_sorted_exp_perf = df.sort_values(by=['Experience', 'Performance Score'], ascending=[False, True])
print(df_sorted_exp_perf.head())
```

```python
print("# Task 33: Create a new DataFrame containing only employees in HR with a Salary above 5000.")
df_hr_salary = df.loc[(df['Department'] == "HR") & (df['Salary'] > 5000)]
print(df_hr_salary)
```

```python
print("# Task 34:  Filter out all rows where the Location is Chicago")
df_no_chicago = df[df['Location'] != "Chicago"]
print(df_no_chicago)
```

```python
print("# Task 35: Delete all rows where the Performance Score is below 2")
df = df[df['Performance Score'] >= 2]
print(df)
```

```python
print("# Task 36:  Find the average Salary for each Department")
avg_salary_dept = df.groupby('Department')['Salary'].mean()
print(avg_salary_dept)
```

```python
print("# Task 37: Calculate the total Salary of all employees grouped by Status")
total_salary_status = df.groupby('Status')['Salary'].sum()
print(total_salary_status)
```

```python
print("# Task 38:  Find the maximum Performance Score for employees in each Location")
max_perf_location = df.groupby('Location')['Performance Score'].max()
print(max_perf_location)
```

```python
print("# Task 39: Calculate the total Salary paid to employees who joined in each year")
if 'Joining Date' in df.columns:
    df['Joining Year'] = pd.to_datetime(df['Joining Date']).dt.year
    total_salary_year = df.groupby('Joining Year')['Salary'].sum()
    print(total_salary_year)
```

```python
print("# Task 40:  Find the department with the highest average Performance Score.")
highest_avg_perf_dept = df.groupby('Department')['Performance Score'].mean().idxmax()
print(f"Department with highest average Performance Score: {highest_avg_perf_dept}")
```

```python
print("# Task 41: Count the number of employees in each Location grouped by their Session")
if 'Session' in df.columns:
    employees_by_location_session = df.groupby(['Location', 'Session']).size()
    print(employees_by_location_session)
```

```python
print("# Task 42: Group employees by Status and calculate the sum, mean, and standard deviation of their Salary")
salary_stats_status = df.groupby('Status')['Salary'].agg(['sum', 'mean', 'std'])
print(salary_stats_status)
```

```python
print("# Task 43: Use the apply() method to create a new column Monthly Salary by dividing the Salary by 12")
df['Monthly Salary'] = df['Salary'].apply(lambda x: x / 12)
print(df[['Salary', 'Monthly Salary']].head())
```

```python
print("# Task 44: Use the applymap() method to convert all string values in the DataFrame to uppercase")
df_uppercase = df.applymap(lambda x: x.upper() if isinstance(x, str) else x)
print(df_uppercase.head())
```

```python
print("# Task 45: Create a second dataset containing Department and Budget. Perform an inner join with the original dataset.")
budget_data = pd.DataFrame({
    'Department': ['Sales', 'IT', 'HR'],
    'Budget': [50000, 30000, 20000]
})
df_inner_join = df.merge(budget_data, on='Department', how='inner')
print(df_inner_join)
```

```python
print("# Task 46: Merge the dataset with another dataset that includes additional information about employee ID and their recent Projects")
projects_list = [i % 5 + 1 for i in range(len(df.index))]
project_data = pd.DataFrame({
    'Employee ID': df.index, 
    'Projects': projects_list  
})
```

```python
print("# Task 47: Perform a left join with a dataset containing Department and Team Lead details, adding Team Lead to the DataFrame")
team_lead_data = pd.DataFrame({
    'Department': ['Sales', 'IT', 'HR'],
    'Team Lead': ['Alice', 'Bob', 'Charlie']
})
df_left_join = df.merge(team_lead_data, on='Department', how='left')
print(df_left_join)
```

```python
print("# Task 48: Create a DataFrame with employee ID and Project Count, then merge it with the original dataset.Calculate the average Salary per project")
import pandas as pd
data = {'Name': ['Cory Escobar', 'Chad Nichols', 'Christine Williams', 'Amber Harris', 'David Olson'],
        'Salary': [80000, 75000, 82000, 78000, 73000]}
df = pd.DataFrame(data)
project_count_data = pd.DataFrame({
    'Employee ID': df.index,
    'Project Count': [3, 2, 5, 4, 1] 
})
df_project_count = df.merge(project_count_data, left_index=True, right_on='Employee ID')
df_project_count['Avg Salary per Project'] = df_project_count['Salary'] / df_project_count['Project Count']

print(df_project_count)
```

```python
print("# Task 49:Calculate the percentage increase in Salary for each employee compared to the average Salary in their Department")
import pandas as pd
data = {
    'Name': ['Cory Escobar', 'Chad Nichols', 'Christine Williams', 'Amber Harris', 'David Olson'],
    'Salary': [80000, 75000, 82000, 78000, 73000],
    'Department': ['HR', 'HR', 'IT', 'IT', 'Finance']  # Added Department column
}
df = pd.DataFrame(data)
df['Dept Avg Salary'] = df.groupby('Department')['Salary'].transform('mean')
df['Salary Increase %'] = ((df['Salary'] - df['Dept Avg Salary']) / df['Dept Avg Salary']) * 100
print(df[['Name', 'Department', 'Salary', 'Dept Avg Salary', 'Salary Increase %']])
```
